```python
from client import run_test
```

```python
# Running Tests
rtt_msg_sizes = [1, 100, 200, 400, 800, 1000]
tput_msg_sizes = [1000, 2000, 4000, 8000, 16000, 32000]

mean_rtts = [run_test("CSA1.bu.edu", 58000, 'rtt', 20, ms, 0) for ms in rtt_msg_sizes]

mean_tputs = [run_test("CSA1.bu.edu", 58000, 'tput', 20, ms, 0) for ms in tput_msg_sizes]

print(f"mean_rtts: {mean_rtts}")
print(f"mean_tputs: {mean_tputs}")
```

```python
mean_rtts_w_delay = [run_test("CSA1.bu.edu", 58000, 'rtt', 20, ms, 2.0) for ms in rtt_msg_sizes]
print(f"mean_rtts_w_delay: {mean_rtts_w_delay}")
```

```python
mean_tputs_w_delay = [run_test("CSA1.bu.edu", 58000, 'tput', 20, ms, 2.0) for ms in tput_msg_sizes]
print(f"mean_tputs_w_delay: {mean_tputs_w_delay}")
```

```python
# Graphing the results
from matplotlib import pyplot as plt

plt.title(f"RTT vs Message Size with no Server Delay")
plt.xlabel("Message Size (bytes)")
plt.ylabel("RTT (ms)")
plt.plot(rtt_msg_sizes, mean_rtts, linestyle='--', marker='o')
plt.show()

plt.title(f"TPUT vs Message Size with no Server Delay")
plt.xlabel("Message Size (bytes)")
plt.ylabel("TPUT (kbps)")
plt.plot(tput_msg_sizes, mean_tputs, linestyle='--', marker='o')
plt.show()
```

```python
plt.title(f"RTT vs Message Size with 2.0s Server Delay")
plt.xlabel("Message Size (bytes)")
plt.ylabel("RTT (ms)")
plt.plot(rtt_msg_sizes, mean_rtts_w_delay, linestyle='--', marker='o')
plt.show()

plt.title(f"TPUT vs Message Size with 2.0s Server Delay")
plt.xlabel("Message Size (bytes)")
plt.ylabel("TPUT (kbps)")
plt.plot(tput_msg_sizes, mean_tputs_w_delay, linestyle='--', marker='o')
plt.show()
```
